```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("student_data.csv")
print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
print(df['age'].mean())
print(df['age'].max())
print(df['G3'].mean())
print(df['G3'].max())
print(df['G3'].min())
print(df.groupby('sex')['G3'].mean())
print(df.groupby('studytime')['G3'].mean())
sns.histplot(df['G3'])
plt.title("Distribution of Final Grades")
plt.show()
sns.boxplot(x='sex', y='G3', data=df)
plt.title("Gender vs Final Grades")
plt.show()
sns.barplot(x='studytime', y='G3', data=df)
plt.title("Study Time vs Grades")
plt.show()
print(df.corr(numeric_only=True)['G3'].sort_values(ascending=False))
# Age shows very weak correlation with final grades (G3), indicating minimal impact.
print(df.groupby('age')['G3'].mean())
# Father's job (Fjob) does not show a strong impact on student performance.
print(df.groupby('Fjob')['G3'].mean())
sns.scatterplot(x='age',y='G3',data=df)
plt.title("Age vs Final grades")
plt.show()
sns.scatterplot(x='absences',y='G3',data=df)
plt.title("Absences vs Final grades")
plt.show()
```
